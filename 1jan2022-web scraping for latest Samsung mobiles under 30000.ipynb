```python
from autoscraper import AutoScraper 

#importing AutoScraper library
```

```python
#passing the url on which we want to work

flipkart_url="https://www.flipkart.com/search?q=latest%20samsung%20%20mobiles%20under%2030000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"

#creating list of elements/sample data that we want to scrape from that page.

wanted_list=["SAMSUNG Galaxy A52 (Awesome Black, 128 GB)","4.4717 Ratings","109 Reviews",
            "64MP + 12MP + 5MP + 5MP | 32MP Front Camera","4500 mAh Lithium-ion Battery","Qualcomm Snapdragon 720G Processor","₹27,999"]
```

```python
#calling the AutoScraper function (inniatialization)
scraper=AutoScraper()

#create object 'scraper.build' which saves data in 'result'.'flipkart_url' and 'wanted_list' are two parameters
result=scraper.build(flipkart_url,wanted_list) 

#prints the result
print(result)        
```

```python
#grouping the similar type of data according to parameters
scraper.get_result_similar(flipkart_url,grouped=True) 
```

```python
# save the rules and search,alias value in key:value pair
scraper.set_rule_aliases({'rule_c6w7':'model_name',
                           'rule_2ois':'review',
                          'rule_8fps':'cam',
                         'rule_mcid':'battery',
                          'rule_jhsc':'processor',
                          'rule_zjv8':'price'}) 

#keep this rule so it can be applied to other pages also
scraper.keep_rules({'rule_c6w7','rule_2ois','rule_8fps','rule_mcid','rule_jhsc','rule_zjv8'})

#save information as give name
scraper.save('flipkart-search')  

result['model_name'] 

# Note rule values will need to be updated every time based on the execution.
```
